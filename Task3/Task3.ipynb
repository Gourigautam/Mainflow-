```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [200, 250, 300, 150, 400, 350],
    'Expenses': [180, 220, 260, 130, 320, 310]
}
```

```python
df=pd.DataFrame(data)
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.bar(df['Month'], df['Sales'], color='b', label='Sales')
```

```python
plt.bar(df['Month'],df['Expenses'], color='r', alpha=0.6, label="Expenses")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
```

```python
# plot the line chart
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='b', label="Sales")
```

```python
plt.plot(df['Month'], df['Expenses'], marker='s', linestyle='--', color='r', label="Expenses")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.title("Monthly Sales and Expenses")
plt.legend()
plt.show()
```
